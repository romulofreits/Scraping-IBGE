```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# Importando o selenium

from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys

servico = Service(ChromeDriverManager().install())

navegador = webdriver.Chrome(service=servico)
```

```python
# Comando para o selenium abrir automaticamente o site passado
navegador.get('https://www.ibge.gov.br/cidades-e-estados/ce.html')
```

```python
# Agora, vamos usar o BeautifulSoup para analisar o código HTML do site e encontrar os links para as páginas dos municípios do Ceará:
soup = BeautifulSoup(navegador.page_source, 'html.parser')
```

```python
type(soup)
```

```python
# A variável links_municipios agora contém os links para as páginas dos municípios do Ceará.

links_municipios = []
for link in soup.find_all('p'):
    href = link.get('href')
    if href is not None and '/municipios/' in href:
        links_municipios.append(href)
```

```python
municipios = []
for link in links_municipios:
    navegador.get(link)
    soup = BeautifulSoup(navegador.page_source, 'html.parser')
    nome = soup.find('p', {'class': 'ind-value'}).text
    pop_estimada = soup.find('p', {'class': 'ind-value'}).find_all('p')[1].text.split(' ')[0].replace('.', '')
    area_km2 = soup.find('div', {'class': 'ind-value'}).find_all('p')[1].text.split(' ')[0].replace(',', '.')
    municipios.append({'nome': nome, 'pop_estimada': pop_estimada, 'area_km2': area_km2})
```

```python
url = "https://www.ibge.gov.br/cidades-e-estados.html?view=municipio"
```

```python
html = requests.get(url).content
print(html)
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
print(soup.prettify())
```

```python
municipios = []
subtitulo = None

for tag in soup.find_all('p'):
    if tag.has_attr('class') and 'subtitulo-tabela' in tag['class']:
        # Se encontrou um subtítulo, salva o texto para referência
        subtitulo = tag.text.strip()
    elif subtitulo and tag.name == 'a':
        # Se encontrou um link de município, adiciona o nome e o link
        municipios.append({'municipio': tag.text.strip(), 'link': tag['href'], 'subtitulo': subtitulo})
        subtitulo = None  # Limpa o subtítulo

print(municipios)
```
